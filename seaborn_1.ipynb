```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = sns.load_dataset("tips")

# Create a new column for tip percentage
data["tip_percentage"] = (data["tip"] / data["total_bill"]) * 100

# Set style
sns.set_style("whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# 1. Scatter plot with regression line
sns.regplot(x="total_bill", y="tip", data=data, ax=axes[0, 0], scatter_kws={"alpha":0.6}, line_kws={"color": "red"})
axes[0, 0].set_title("Total Bill vs Tip with Regression Line")

# 2. Scatter plot with gender-based coloring
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=data, ax=axes[0, 1], palette="coolwarm")
axes[0, 1].set_title("Total Bill vs Tip (Colored by Gender)")

# 3. Histogram of Tip Percentage
sns.histplot(data["tip_percentage"], bins=20, kde=True, ax=axes[1, 0], color="purple")
axes[1, 0].set_title("Distribution of Tip Percentage")

# 4. Boxplot to analyze tipping behavior by gender
sns.boxplot(x="sex", y="tip", data=data, ax=axes[1, 1], palette="coolwarm")
axes[1, 1].set_title("Tip Amount by Gender")

# Adjust layout for better spacing
plt.tight_layout()

# Show the plots
plt.show()

import os
os.getcwd()


```
