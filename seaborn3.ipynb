```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the built-in 'tips' dataset from seaborn
data = sns.load_dataset("tips")

# Display first 5 rows to understand the dataset
print(data.head())

# Set the visual style
sns.set_style("whitegrid")

# 1. Histogram of total bill amounts
plt.figure(figsize=(8, 5))
sns.histplot(data['total_bill'], kde=True, color='skyblue')
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill Amount")
plt.ylabel("Frequency")
plt.show()

# 2. Boxplot of tip amount by gender
plt.figure(figsize=(8, 5))
sns.boxplot(x='sex', y='tip', data=data, palette='pastel')
plt.title("Tip Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Tip Amount")
plt.show()

# 3. Heatmap of correlations between numeric variables
plt.figure(figsize=(8, 6))

# Select only numeric columns before calculating correlation
correlation = data.select_dtypes(include='number').corr()

sns.heatmap(correlation, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Correlation Between Numeric Features")
plt.show()
```
